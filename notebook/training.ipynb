Preprocessing

```python
df["gender"] = df["gender"].map({"Female" : 1, "Male" : 0})
mapped = df[["Partner","Dependents","PhoneService","PaperlessBilling","Churn"]]

for col in mapped:
    df[col] = df[col].map({"Yes" : 1, "No" : 0})
```

```python
for col in df.columns:
    print(f"Columns: {col}")
    print(df[col].unique())
```

```python
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
```

```python
df["TotalCharges"].isna().sum()
```

```python
df["TotalCharges"] = df["TotalCharges"].fillna(df["TotalCharges"].median())
df["TotalCharges"].isna().sum()
```

```python
print(df["gender"])
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split

X = df.iloc[:,0:-1]
y = df.iloc[:,-1]

X_train, X_test, y_train, y_test = train_test_split(X,y,stratify=y,test_size=0.2,random_state=42)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler

cat = ["MultipleLines","InternetService","OnlineSecurity","OnlineBackup",
          "DeviceProtection","TechSupport","StreamingTV","StreamingMovies",
          "Contract","PaymentMethod"]

num = ["tenure", "MonthlyCharges", "TotalCharges"]

preprocessor = ColumnTransformer(
    transformers=
    [("cat",OneHotEncoder(handle_unknown="ignore", sparse_output=False), cat),
     ("num",StandardScaler(), num)
     ],
    remainder="passthrough"
)
```

```python
from sklearn.linear_model import LogisticRegression

model_logres = LogisticRegression(class_weight="balanced")

pipe_logres = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("model_logres", model_logres)
])
```

```python
from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier(class_weight="balanced")

pipe_rf = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("model_ranfor", model_rf)
])
```

```python
scale_pos_weight = (y_train == 0).sum() / (y_train == 1).sum()
```

```python
from xgboost import XGBClassifier

model_xgb = XGBClassifier(
    random_state=42,
    scale_pos_weight=scale_pos_weight,
    objective="binary:logistic",
    eval_metric="logloss"
)

pipe_xgb = Pipeline(steps=[
    ("preprocess", preprocessor),
    ("model", model_xgb)
])
```

```python
from lightgbm import LGBMClassifier

model_lgbm = LGBMClassifier(class_weight="balanced", verbose = -1)

pipe_lgbm = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("model_lgbm", model_lgbm)
])
```

```python
from catboost import CatBoostClassifier

model_cat = CatBoostClassifier(verbose=0)

pipe_cat = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("model_cat", model_cat)
])
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

models = {
    "Logres" : pipe_logres,
    "RandomForest" : pipe_rf,
    "XGB" : pipe_xgb,
    "LightGBM" : pipe_lgbm,
    "CatBoost" : pipe_cat
}

results = []

for name, model in models.items():
    
    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_test)
    
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    rec = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    
    results.append({
        "Model": name,
        "Accuracy": acc,
        "Precision": prec,
        "Recall": rec,
        "F1 Score": f1
    })

results_df = pd.DataFrame(results)

results_df = results_df.sort_values(by="Recall", ascending=False)

print(results_df)
    
```

Cross Validation 

```python
from sklearn.model_selection import StratifiedKFold, cross_validate

skf = StratifiedKFold(n_splits=5, shuffle=True,random_state=42)

model = {
    "Logres" : pipe_logres,
    "LightGBM" : pipe_lgbm,
    "XGBoost" : pipe_xgb
}

cv_results = {}

for name, model in models.items():
    scores = cross_validate(
        model,
        X,
        y,
        cv = skf,
        scoring=["recall", "precision", "f1"],
        return_train_score=False
    )
    
    cv_results[name] = {
        "Recall" : scores["test_recall"].mean(),
        "Precison" : scores["test_precision"].mean(),
        "F1" : scores["test_f1"].mean()
    }
    
    cv_df = pd.DataFrame(cv_results).T
    print(cv_df.sort_values(by="Recall", ascending=False)
)
    
```

Threshold Tuning

```python
models = {
    "LogReg": pipe_logres,
    "LightGBM": pipe_lgbm
}

threshold = 0.3
results = []

for name, model in models.items():
    
    y_prob_thres = model.predict_proba(X_test)[:, 1]
    y_pred_thres = (y_prob_thres > threshold).astype(int)
    
    precision = precision_score(y_test, y_pred_thres)
    recall = recall_score(y_test, y_pred_thres)
    f1 = f1_score(y_test, y_pred_thres)
    
    tn, fp, fn, tp = confusion_matrix(y_test, y_pred_thres).ravel()
    
    results.append({
        "Model": name,
        "Threshold": threshold,
        "Precision": precision,
        "Recall": recall,
        "F1 Score": f1,
        "FN (Missed Churn)": fn,
        "FP (False Alarm)": fp
    })

results_df = pd.DataFrame(results)

print(results_df)
```
